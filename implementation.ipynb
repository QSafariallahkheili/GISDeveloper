```python
import pandas as pd # data processing
import numpy as np # working with arrays
from sklearn.model_selection import train_test_split # data split
from sklearn.linear_model import LinearRegression # OLS algorithm

import pickle
import folium
import mapclassify

df = pd.read_csv('validation.csv')
df['price'] = df['price'].astype('int64')
df['bathrooms'] = df['bathrooms'].astype('int64')
df['floors'] = df['floors'].astype('int64')
# to estimate building age: 2021 - yr_built
df['yr_built'] =  2021 - df['yr_built']
x_var = df[['sqft_living', 'sqft_above', 'sqft_living15', 'bathrooms']].values

#loading the developed machine learning model
model = pickle.load(open("ols.pkl", "rb"))

#price prediction 
y_pred = []
for i in range (len(df.index)):
    y_pred.append(model.predict([x_var[i]]))
    #print(y_pred[i][0][0])


locations = df[['lat', 'long']].values
atts = df[['id','price']].values

intervals = mapclassify.Quantiles(df[['price']].values, k=5)


# apply predicted price on border color and real price on fill color

def styleFill(x):
    fillcolor='#FF0000'
    if x<354340.00:
        fillcolor='#ffffd4'
    elif x>=354340.00 and x<444600.00: 
        fillcolor = '#fed98e'
    elif x>=444600.00 and x<549996.00: 
        fillcolor = '#fe9929'
    elif x>=549996.00 and x<759000.00:
        fillcolor = '#d95f0e'
    else:
        fillcolor = '#993404'
    return fillcolor

def styleBorder(x):
    bordercolor='#FF0000'
    if x<354340.00:
        bordercolor='#ffffd4'
    elif x>=354340.00 and x<444600.00: 
        bordercolor = '#fed98e'
    elif x>=444600.00 and x<549996.00: 
        bordercolor = '#fe9929'
    elif x>=549996.00 and x<759000.00:
        bordercolor = '#d95f0e'
    else:
        bordercolor = '#993404'
    return bordercolor

m = folium.Map(location=[47.5162, -121.87200000000001])
for i in range (len(df.index)):
    folium.CircleMarker(
        location=[locations[i][0],locations[i][1]],
        radius = 5,
        color=styleBorder(y_pred[i][0][0]),
        opacity= 1,
        fill_color=styleFill(atts[i][1]),
        fill_opacity= 1,
        popup='id: {} </br> Price: {} </br> Predicted Price: {}'.format(atts[i][0], atts[i][1], y_pred[i][0][0]),
        #icon=folium.Icon(icon="cloud"),
    ).add_to(m)
    
    

m
```
